<h1>Задача 1</h1>

```python
def reverse_mas(arr, st, fin):
    while st < fin:
        arr[st], arr[fin] = arr[fin], arr[st]
        st += 1
        fin -= 1
    return arr
    
def reverse_k(arr, k):
    n = len(arr)
    k = k % n
    arr = reverse_mas(arr, 0, n-1)
    arr = reverse_mas(arr, 0, k-1)
    arr = reverse_mas(arr, k, n-1)
    return arr
```

Полный реверс массива происходит за О(n) время, мы его делаем суммарно два раза, но это все равно О(n)

<h1>Задача 2</h1>

```python
def concatenate(arr1, arr2):
    res = []
    n, m = len(arr1), len(arr2)
    p1, p2 = 0, 0
    while p1 < n or p2 < m:
        if p1 < n and p2 < m:
            n1, n2 = arr1[p1], arr2[p2]
            if n1 < n2:
                res.append(n1)
                p1 += 1
            else:
                res.append(n2)
                p2 += 1
        elif p1 < n:
            res.append(arr1[p1])
            p1 += 1
        else:
            res.append(arr2[p2])
            p2 += 1
    return res
```

<h1>Задача 3</h1>

```python
def merge(arr1, arr2):
    pos = len(arr1)-1
    p1, p2 = pos-len(arr2), len(arr2)-1
    while p1 >= 0 or p2 >= 0:
        if p1 >= 0 and p2 >= 0:
            if arr1[p1] > arr2[p2]:
                arr1[pos] = arr1[p1]
                p1 -= 1
            else:
                arr1[pos] = arr2[p2]
                p2 -= 1
        elif p1 >= 0:
            arr1[pos] = arr1[p1]
            p1 -= 1
        else:
            arr1[pos] = arr2[p2]
            p2 -= 1
        pos -= 1
```

<h1>Задача 4</h1>

```python
def sort_01(arr):
    st, fin = 0, len(arr)-1
    while st < fin:
        if arr[st] == 0:
            st += 1
        elif arr[fin] == 1:
            fin -= 1
        else:
            arr[st], arr[fin] = arr[fin], arr[st]
            st += 1
            fin -= 1
```

<h1>Задача 5</h1>

```python
def sort_012(arr):
    st, mid, fin = 0, 0, len(arr)-1
    while mid <= fin:
        if arr[mid] == 0:
            arr[mid], arr[st] = arr[st], arr[mid]
            st += 1
            mid += 1
        elif arr[mid] == 1:
            mid += 1
        elif arr[mid] == 2:
            arr[mid], arr[fin] = arr[fin], arr[mid]
            fin -= 1
```

<h1>Задача 6</h1>

```python
def replace_even(arr):
    pos_cur = 0
    for i in range(len(arr)):
        if arr[i] % 2 == 0:
            arr[i], arr[pos_cur] = arr[pos_cur], arr[i]
            pos_cur += 1
```

```python
def zeros_to_end(arr):
    st, fin = 0, len(arr)-1
    while st < fin:
        if arr[fin] == 0:
            fin-=1
        elif arr[st] == 0:
            arr[fin], arr[st] = arr[st], arr[fin]
            st += 1
            fin -= 1
        else:
            st += 1
```
